# Graph: the Curves


- Read 'unrate.csv' in a dataframe and assign this result to the variable unrate.
- Use the pandas.to_datetime() function to convert the 'DATE' column to datetime values.
- Display the first 12 unrate values corresponding to a whole year.


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec
import datetime as dt
# Read 'unrate.csv' in a dataframe and assign this result to the variable unrate.
unrate = pd.read_csv("unrate.csv")
# Use the pandas.to_datetime() function to convert the 'DATE' column to datetime values.
unrate["DATE"]=pd.to_datetime(unrate["DATE"])
# Display the first 12 unrate values corresponding to a whole year.
unrate.head(12)
```

## Adding data to our charts


- Generate a line graph to visualize the unemployment rates for 1948:
>- Take for x the first 12 values in the column 'DATE'.
>- Take for y the first 12 values of the column 'UNRATE'.
- Display the graph.


```python
# Generate a line graph to visualize the unemployment rates for 1948:
Date=unrate["DATE"]
Rate=unrate["UNRATE"]
x=Date.head(12)
y=Rate.head(12)
plt.plot(x,y, color='green', linewidth=2)
plt.title("Line graph to visualize the unemployment rates for 1948")  
plt.ylabel('Unrate')  
plt.xlabel('Date')
# Display the graph.
plt.show() 
```

## Superimposing curves


- Set the display area to 6 inches wide and 3 inches high.
- Generate 2 curves in a single graph using the 'MONTH' column for the x-axis instead of the 'DATE' column:
>- One curve with 1948 values in 'red' color.
>- A curve with the 1949 values of color 'blue'.
- Use plt.show() to display the curves.


```python
# Set the display area to 6 inches wide and 3 inches high.
fig= plt.figure(figsize=(6,3))
# Generate 2 curves in a single graph using the 'MONTH' column for the x-axis instead of the 'DATE' column:
Month=pd.Series(unrate["DATE"].dt.month)
x=Month[0:12]
y=Rate[0:12]
# One curve with 1948 values in 'red' color.
plt.plot(x,y, color='red', linewidth=2)
x1=Month[12:24]
y1=Rate[12:24]
# A curve with the 1949 values of color 'blue'.
plt.plot(x1,y1, color='blue', linewidth=2)
plt.ylabel('Unemployment Rate (%)')  
plt.xlabel('Month')
plt.show()
```

## Add more curves


- Set the graph display area to 10 inches wide and 6 inches high.
- Generate the following curves on the same graph:
>- 1948: red curve 'red'.
>- 1949: curve in 'blue' blue
>- 1950: curved in green
>- 1951: curve in orange 'orange'.
>- 1952: curved in 'black' black
- Use plt.show() to display the curves.


```python
# Set the graph display area to 10 inches wide and 6 inches high.
fig= plt.figure(figsize=(10,6))
Month=pd.Series(unrate["DATE"].dt.month)
x=Month[0:12]
y=Rate[0:12]
# 1948: red curve 'red'.
plt.plot(x,y, color='red', linewidth=2)
x1=Month[12:24]
y1=Rate[12:24]
# 1949: curve in 'blue' blue
plt.plot(x1,y1, color='blue', linewidth=2)
x2=Month[24:36]
y2=Rate[24:36]
# 1950: curved in green
plt.plot(x2,y2, color='green', linewidth=2)
x3=Month[36:48]
y3=Rate[36:48]
# 1951: curve in orange 'orange'.
plt.plot(x3,y3, color='orange', linewidth=2)
x4=Month[48:60]
y4=Rate[48:60]
# 1952: curved in 'black' black
plt.plot(x4,y4, color='black', linewidth=2)
plt.ylabel('Unemployment Rate (%)')  
plt.xlabel('Month')
plt.show()
```

## Add a caption


- Modify the previous code to include a legend specifying the year for each colour.
>- This means that the 1948 data curve using the red colour 'red' will have the legend '1948' as its title.
- Position the legend in the upper left corner 'upper left'.
- Display the graph.


```python
# Modify the previous code to include a legend specifying the year for each colour.
fig= plt.figure(figsize=(10,6))
ax = plt.subplot()
Month=pd.Series(unrate["DATE"].dt.month)
x=Month[0:12]
y=Rate[0:12]
ax.plot(x,y, color='red', linewidth=2, label='1948')
x1=Month[12:24]
y1=Rate[12:24]
ax.plot(x1,y1, color='blue', linewidth=2, label='1949')
x2=Month[24:36]
y2=Rate[24:36]
ax.plot(x2,y2, color='green', linewidth=2, label='1950')
x3=Month[36:48]
y3=Rate[36:48]
ax.plot(x3,y3, color='orange', linewidth=2, label='1951')
x4=Month[48:60]
y4=Rate[48:60]
ax.plot(x4,y4, color='black', linewidth=2, label='1952')
plt.ylabel('Unemployment Rate (%)')  
plt.xlabel('Month')
# Position the legend in the upper left corner 'upper left'.
ax.legend(loc='upper left')
# Display the graph.
plt.show()
```
